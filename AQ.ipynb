# Algorithmic question

In this algorithmic question we have to unlock a smartphone, that is given a 3\*3 grid and the direction of each line segment find how many possible patterns could unlock. A great help: the pattern cannot intersect with itself.

The first point is to check if Alex has glimpsed well the directions 

```python
def con_ord(stringa):
    
    """
    function which check if 'stringa' has the structure we want.
    """
    
    o = ['R','L']  # orizzontal letter
    v = ['U','D']  # vertical letter
    
    i = 0
    while i < len(stringa)-1 and i != -1:  # check if the two type of letters take turns
        if (stringa[i] in o and stringa[i+1] in v) or (stringa[i] in v and stringa[i+1] in o):
            i += 1
        else: 
            i = -1
            
    return i
```

Now given a point in the 3\*3 grid find how many patterns there are, indicating with $[0,0],[0,1],[0,2],[1,0],[1,1],[1,2],[2,0],[2,1],[2,2]$ the different point, in which $[0,0]$ is the point 
lower left, and the first item represent the orizzontal side and the second item the vertical one.

```python
def find_one(stringa, ora, passati):
   
    """
    
    function which finds how many patterns from 'ora' and given 'passati' the 
    points already passed by the pattern:
    
    stringa: str(), word=pattern;
    ora: [int,int], starting point;
    passati: list(), points already passed by the pattern.
    
    """
    
    s = 0
    
    if min(ora)<0 or max(ora)>2 or (ora in passati): # if the pattern goes out from the grid => stop
        pass
    
    else:
        passati.append(ora)     # append the starting point in passati
        if stringa == '':       # if stringa == '', we have found a pattern
            s += 1
        else:
            pas1 = passati.copy()   # copy twice 'passati'
            pas2 = passati.copy()
            ora1 = ora.copy()       # copy starting point
            
            if stringa[0] == 'R':      # if the first letter in 'stringa' is 'R', go right
                ora1[0] = ora1[0] + 1  # only one step
                s += find_one(stringa[1:], ora1, pas1) # find the patterns after this step
                ora1[0] = ora1[0] + 1  # two steps
                s += find_one(stringa[1:], ora1, pas2) # find the patterns after this step
                
            elif stringa[0] == 'L':
                ora1[0] = ora1[0] - 1
                s += find_one(stringa[1:], ora1, pas1)
                ora1[0] = ora1[0] - 1
                s += find_one(stringa[1:], ora1, pas2)
            
            elif stringa[0] == 'U':
                ora1[1] = ora1[1] + 1
                s += find_one(stringa[1:], ora1, pas1)
                ora1[1] = ora1[1] + 1
                s += find_one(stringa[1:], ora1, pas2)
            
            else:
                ora1[1] = ora1[1] - 1
                s += find_one(stringa[1:], ora1, pas1)
                ora1[1] = ora1[1] - 1
                s += find_one(stringa[1:], ora1, pas2)
    
    return s    # return the number of patterns 
```

Finally put the two previous functions together: check if the directions seen by Alex indicate a possible pattern, and find the number of patterns from each grid point.

```python
def find_all(stringa):
    
    """
    function which counts how many total patterns 
    """
    
    pino = con_ord(stringa)   # is 'stringa' a good pattern?
    if pino == -1:           
        return print("Alex didn't glimps well...")
    else:
        punti = [[0,0],[0,1],[0,2],[1,0],[1,1],[1,2],[2,0],[2,1],[2,2]]
        totale = 0
        for i in punti:  # sum the patterns found starting from each grid point
            totale += find_one(stringa, i, [])
    return totale
```
