```python
def trovapath(stringa, ora, passati):

    s = 0
    
    if min(ora)<0 or max(ora)>2 or (ora in passati):
        pass
    
    else:
        passati.append(ora)
        if stringa == '':
            s += 1
        else:
            pas1 = passati.copy()
            pas2 = passati.copy()
            ora1 = ora.copy()
            
            if stringa[0] == 'R':
                ora1[0] = ora1[0] + 1
                s += trovapath(stringa[1:], ora1, pas1)
                ora1[0] = ora1[0] + 1
                s += trovapath(stringa[1:], ora1, pas2)
                
            elif stringa[0] == 'L':
                ora1[0] = ora1[0] - 1
                s += trovapath(stringa[1:], ora1, pas1)
                ora1[0] = ora1[0] - 1
                s += trovapath(stringa[1:], ora1, pas2)
            
            elif stringa[0] == 'U':
                ora1[1] = ora1[1] + 1
                s += trovapath(stringa[1:], ora1, pas1)
                ora1[1] = ora1[1] + 1
                s += trovapath(stringa[1:], ora1, pas2)
            
            else:
                ora1[1] = ora1[1] - 1
                s += trovapath(stringa[1:], ora1, pas1)
                ora1[1] = ora1[1] - 1
                s += trovapath(stringa[1:], ora1, pas2)
    
    return s


def conord(stringa):
    o = ['R','L']
    v = ['U','D']
    
    i = 0
    while i < len(stringa)-1 and i != -1:
        if (stringa[i] in o and stringa[i+1] in v) or (stringa[i] in v and stringa[i+1] in o):
            i += 1
        else: 
            i = -1
            
    return i
    

def cosabuona(stringa):
    pino = conord(stringa)
    if pino == -1:
        return print('Non Ã¨ buona')
    else:
        punti = [[0,0],[0,1],[0,2],[1,0],[1,1],[1,2],[2,0],[2,1],[2,2]]
        totale = 0
        for i in punti:
            totale += trovapath(stringa, i, [])
    return totale
```
